## Cambios por CLAUDE

```python
from langchain.prompts import ChatPromptTemplate

medical_prompt = ChatPromptTemplate.from_template("""
Eres un asistente médico especializado en la gestión de historiales de prescripciones médicas.
Tu función es analizar y presentar información de recetas médicas anteriores de manera clara y organizada.

Instrucciones específicas:
- Organiza la información cronológicamente
- Incluye fechas de prescripción
- Lista los medicamentos con sus dosis y frecuencias
- Menciona los diagnósticos asociados
- Incluye el nombre del médico tratante
- Mantén la confidencialidad del paciente
- Si la información solicitada no está en el contexto, indícalo claramente

Contexto del historial médico:
{context}

Pregunta: {question}

Respuesta: Basándome en el historial de prescripciones médicas disponible:
""")

# Función mejorada para formatear documentos
def format_docs(docs):
    # Ordenar documentos por fecha (asumiendo que existe un metadata['date'])
    sorted_docs = sorted(docs, key=lambda x: x.metadata.get('date', ''), reverse=True)
    return "\n\n--- Nueva Receta ---\n".join(doc.page_content for doc in sorted_docs)

# Configuración del retriever con parámetros optimizados
def setup_retriever(vectorstore):
    return vectorstore.as_retriever(
        search_type="similarity",
        search_kwargs={
            "k": 10,  # Aumentado para asegurar que recuperamos todas las recetas relevantes
            #"score_threshold": 0.5  # Umbral de similitud
        }
    )

# Función principal de configuración RAG
def setup_medical_rag_chain(retriever, llm):
    rag_chain = (
        {
            "context": retriever | format_docs,
            "question": RunnablePassthrough()
        }
        | medical_prompt
        | llm
        | StrOutputParser()
    )
    return rag_chain

# Ejemplo de procesamiento de consultas comunes
ejemplo_consultas = [
    "¿Qué medicamentos se le recetaron al paciente en su última visita?",
    "¿Cuál es el historial completo de prescripciones del paciente?",
    "¿Qué diagnósticos ha recibido el paciente?",
    "¿Quiénes han sido los médicos tratantes del paciente?",
    "¿Cuál fue la última dosis prescrita de [medicamento específico]?"
]

# Configuración del LLM con parámetros específicos para uso médico
def initialize_medical_llm(api_key):
    return ChatGoogleGenerativeAI(
        model="gemini-1.5-flash",
        google_api_key=api_key,
        temperature=0.3,  # Temperatura baja para respuestas más precisas
        max_tokens=1024,  # Ajustar según necesidad
        streaming=True
    )

# Create vector store
def create_vectorstore(splits, embeddings):
    return Chroma.from_documents(
        documents=splits,
        embedding=embeddings,
        persist_directory="./chroma_db"  # Add persistence
    )


# Initialize embeddings
def initialize_embeddings():
    return HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-mpnet-base-v2",
        model_kwargs={'device': 'cpu'}  # Explicitly specify device
    )
```

```python
from datetime import datetime
from typing import List, Dict

class MedicalPrescriptionProcessor:
    def __init__(self):
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=1000,  # Tamaño mayor para mantener contexto médico
            chunk_overlap=200,  # Mayor superposición para no perder información
            separators=["\n\n", "\n", ". ", " ", ""],
            length_function=len
        )

    def process_prescription(self, file_path: str) -> List[Dict]:
        loader = PyMuPDFLoader(file_path)
        pages = loader.load()

        # Extraer fecha del documento (esto necesitaría adaptarse a tu formato específico)
        try:
            date = self._extract_date(pages[0].page_content)
        except:
            date = datetime.now().strftime("%Y-%m-%d")

        # Agregar metadata relevante
        for page in pages:
            page.metadata.update({
                "date": date,
                "type": "prescription",
                "source": file_path
            })

        # Dividir manteniendo estructura de prescripción
        splits = self.text_splitter.split_documents(pages)
        return splits

    def _extract_date(self, text: str) -> str:
        # Implementar extracción de fecha según el formato de tus documentos
        # Este es un placeholder
        return datetime.now().strftime("%Y-%m-%d")

# Ejemplo de uso
def main():
    prescription_files = [
        "/content/drive/MyDrive/Insulino/receta-medica-1102470067 (1).pdf",
        "/content/drive/MyDrive/Insulino/receta-medica-1102470067.pdf"
    ]

    # Inicializar componentes
    processor = MedicalPrescriptionProcessor()
    llm = initialize_medical_llm(GOOGLE_API_KEY)
    embeddings = initialize_embeddings()

    # Procesar documentos
    all_splits = []
    for file in prescription_files:
        splits = processor.process_prescription(file)
        all_splits.extend(splits)

    # Crear vectorstore y retriever
    vectorstore = create_vectorstore(all_splits, embeddings)
    retriever = setup_retriever(vectorstore)

    # Configurar cadena RAG
    rag_chain = setup_medical_rag_chain(retriever, llm)

    # Ejemplo de consulta
    query = "Por favor, muestra todo el historial de prescripciones del paciente 1102470067, ordenado por fecha"
    for chunk in rag_chain.stream(query):
        print(chunk, end="", flush=True)
```
